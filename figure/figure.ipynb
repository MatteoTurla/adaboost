```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = (15, 10)
```

```python
x = np.linspace(0.001, 0.5)
y = 0.5*np.log((1-x)/x)
```

```python
plt.plot(x,y)
plt.xlabel("error")
plt.ylabel("weight")
```

```python
gamma_1 = 0.001
gamma_2 = 0.01
gamma_3 = 0.1
x = np.linspace(1, 1000, num=1000)
y1 = np.exp(-2*x*gamma_1)
y2 = np.exp(-2*x*gamma_2)
y3 = np.exp(-2*x*gamma_3)

plt.plot(x, y1, label = "gamma = 0.001")
plt.plot(x, y2, label = "gamma = 0.01")
plt.plot(x, y3, label = "gamma = 0.1")
plt.xlabel("boosting rounds")
plt.ylabel("training error")
plt.legend()
plt.savefig("t.png")
```
